```python
# Calculations and Condition
num = int(input("Enter an Integer: "))

div_3 = (num % 3 == 0)
div_5 = (num % 5 == 0) 

if div_3 and div_5:
    print("Divisible by 3 and 5.")
elif (div_3 or div_5) and not (div_3 and div_5):
    print("Divisible by only one.")
elif not div_3 and not div_5:
    print("Divisible by none.")
```

```python
# Reverse a string
word = "Saniya"

char_list  = list(word)

char_list.reverse()

reverse_word = "".join(char_list)

print("String: {} | Reversed String: {}".format(word, reverse_word))
```

```python
# String Counting
text = "dadsgvCTFGDAD12345698876312"

upper_count = 0
lower_count = 0
digit_count = 0

for char in text:
    if char.isupper():
        upper_count += 1
    elif char.islower():
        lower_count += 1
    elif char.isdigit():
        digit_count += 1

print("Upper Count : {} | Lower Count: {} | Digits: {}".format(upper_count, lower_count, digit_count))
```
