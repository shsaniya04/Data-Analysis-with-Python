```python
# String Counting
text = "dadsgvCTFGDAD12345698876312"

upper_count = 0
lower_count = 0
digit_count = 0

for char in text:
    if char.isupper():
        upper_count += 1
    elif char.islower():
        lower_count += 1
    elif char.isdigit():
        digit_count += 1

print("Upper Count : {} | Lower Count: {} | Digits: {}".format(upper_count, lower_count, digit_count))
```

```python
# Twitter Sentiment Analysis

tweets = [
    "Wow, what a great day today!! #sunshine",
    "I feel sad about the things going on around us. #covid19",
    "I'm really excited to learn Python with @JovianML #zerotopandas",
    "This is a really nice song. #linkinpark",
    "The python programming language is useful for data science",
    "Why do bad things happen to me?",
    "Apple announces the release of the new iPhone 12. Fans are excited.",
    "Spent my day with family!! #happy",
    "Check out my blog post on common string operations in Python. #zerotopandas",
    "Freecodecamp has great coding tutorials. #skillup"
]
```

```python
no_of_tweets = len(tweets)

print("No of Tweets: ", no_of_tweets)
```

```python
happy_words = ['great', 'excited', 'nice', 'happy', 'wonderful', 'amazing', 'good', 'best']
sad_words = ['sad', 'bad', 'tragic', 'unhappy', 'worst']
```

```python
sample_tweet = tweets[0]
sample_tweet
```

```python
is_tweet_happy = False

for word in happy_words: 
    if word in sample_tweet:
        is_tweet_happy = True
```

```python
is_tweet_happy
```

```python
no_of_happy_tweets = 0

for tweet in tweets:
    for happy in happy_words:
        if happy in tweet:
            no_of_happy_tweets += 1
            break

print("Number of happy tweets:", no_of_happy_tweets)
```

```python
happy_fraction = no_of_happy_tweets / len(tweets)

print("The fraction of happy tweets is:", happy_fraction)
```

```python
no_of_sad_tweets = 0

for tweet in tweets:
    for sad in sad_words:
        if happy in tweet:
            no_of_sad_tweets += 1
            break

print("Number of happy tweets:", no_of_sad_tweets)
```

```python
sad_fraction = no_of_sad_tweets / len(tweets)

print("The fraction of sad tweets is:", sad_fraction)
```

```python
sentiment_score = happy_fraction - sad_fraction
print("The sentiment score for the given tweets is", sentiment_score)
```
