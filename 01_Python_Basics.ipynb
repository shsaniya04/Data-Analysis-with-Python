## Arithmetic Operation
### Python Operators 

| Operator   | Purpose           | Example     | Result    |
|------------|-------------------|-------------|-----------|
| `+`        | Addition          | `2 + 3`     | `5`       |
| `-`        | Subtraction       | `3 - 2`     | `1`       |
| `*`        | Multiplication    | `8 * 12`    | `96`      |
| `/`        | Division          | `100 / 7`   | `14.28..` |
| `//`       | Floor Division    | `100 // 7`  | `14`      |    
| `%`        | Modulus/Remainder | `100 % 7`   | `2`       |
| `**`       | Exponent          | `5 ** 3`    | `125`     |


```python
5 + 4 + 2 + 7
```

```python
100 - 73
```

```python
27 * 2
```

```python
100 / 7
```

```python
100 // 7
```

```python
100 % 7
```

As you might expect, certain operators like `/`and `*` take precendence over other operators like `+` and `-` as per mathematical conventions. You can use parantheses i.e. `(` and `)` to specify the order in which operations are performed.

```python
((5 * 4) + (10 // 3)) * ((234 - 43) / ( 98 + 2))
```

```python
# Store input data variables
cost_of_ice_bag = 1.25
profit_margin = .2
no_of_bags = 500

# Perform calculations
profit_per_bag = profit_margin * cost_of_ice_bag
total_profit = profit_per_bag * no_of_bags

# Display result
print("The grocery store makes a total profit of $", total_profit)
```

```python
# Store input data variables
population_of_town = 198568
no_of_men = 45312
no_of_women = 35678

# Perform calculations
total_adults = no_of_men + no_of_women
no_of_children = population_of_town - total_adults

# Display result
print("The total no of children present in town is ", no_of_children)
```

```python
# Store input data variables
no_of_boxes = 2425
pencil_per_box = 24

# Perform calculations
total_pencil = no_of_boxes * pencil_per_box

# Display result
print("The total no of pencil do all the boxes have in all is ", total_pencil)
```

## Logical Operators
### Combining conditions with logical operators

The logical operators `and`, `or` and `not` operate upon conditions and `True` & `False` values (also known as *booleans*). `and` and `or` operate on two conditions, whereas `not` operates on a single condition.

```python
# AND
high_no > 0 and low_no < 8
```

```python
# OR
medium_no > 30 or low_no > 6
```

```python
# NOT
not medium_no != 27
```
